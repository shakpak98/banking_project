```python
from tkinter import*
from tkinter.ttk import Combobox
from PIL import Image,ImageTk
from datetime import datetime
from tkinter import messagebox
```

```python
import sqlite3
con=sqlite3.connect(database='illuminati_bank.sqlite')
cur=con.cursor()
table1='create table accounts(account_no integer primary key autoincrement,account_name text,account_password text,account_email text,account_mob text,account_type text,account_bal float,account_opendate text)'
table2='create table txn(txn_account integer,txn_amount text,txn_updated_balance float,txn_date text,txn_type text)'


try:
    cur.execute(table1)
    cur.execute(table2)
    print('tables created')
except:
    print('something went wrong in database, Table(s) might already exist')
con.commit()
con.close()
```
